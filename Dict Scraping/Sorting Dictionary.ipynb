```python
import json
```

```python
with open('./bigFreqList.json') as infile:
    freqList = json.load(infile)
```

```python
sortedList = sorted(freqList.items(), key=lambda x: x[1], reverse=True)
```

```python
sortedList[13000:14000]
```

```python
dictionary = {}
dictionary['data'] = sortedList
```

```python
with open('./sortedFreqList.json', 'w') as outfile:
    json.dump(dictionary, outfile)
```

```python
import json
import re
```

```python
#got from https://medium.com/the-artificial-impostor/detecting-chinese-characters-in-unicode-strings-4ac839ba313a
#detect whether there is Han character in word
def detectCJK(word):
    if re.search("[\u4e00-\u9FFF]", word):
        return True
    return False
```

```python
detectCJK("洲南極")
```

```python
detectCJK("asd")
```

```python
#find whether the word is in the list
def containsWord(word, lst):
    return word.lower() in [elem.lower() for elem in lst]
```

```python
containsWord("abC", ["abc"])
```

```python
containsWord("abC", ["abd"])
```

```python
#generating a code1-code3 dictionary through the dictionaries generated by triangulating code1-code2-code3
#and code1-code4-code3
def confirmTriangulation(code1, code2, code3, code4):
    print("Confirming from ")
    code123_path = "./Triangulate/" + code1 + "-" + code2 + "-" + code3 + ".json"
    with open(code123_path, "r") as infile:
        code123 = json.load(infile)
    print(code123_path, "with", len(code123), "items")
    code143_path = "./Triangulate/" + code1 + "-" + code4 + "-" + code3 + ".json"
    with open(code143_path, "r") as infile:
        code143 = json.load(infile)
    print(code143_path, "with", len(code143), "items")
    
    code13 = {}
    for (word, translations) in code123.items():
        #the word is also in the other dict
        if (word in code143):
            for translation in translations: 
                
                #remove weird square brackets from wikt2dict
                translation = translation.replace("[","").replace("]","")
                
                #ignoring Chinese characters
                if detectCJK(translation): #detect whether there are any Chinese characters in there
                    continue
                    
                #the translation is also in the other dict for the word
                elif(containsWord(translation, code143[word])):
                    #the word is not in our dict yet
                    if (word not in code13):
                        code13[word] = [translation]
                    
                    #that particular translation is not in our dict yet
                    elif(not containsWord(translation, code13[word])):
                        code13[word].append(translation)
    
    output_path = "./Confirmed/" + code1 + "-" + code3 + ".json"
    with open(output_path, "w") as outfile:
        json.dump(code13, outfile)
    print("Resulted in ")
    print(output_path, "with", len(code13), "items")
```

```python
confirmTriangulation("hu","en","vi","fr")
```

```python
import glob
import nltk
import string
nltk.download('punkt')
import json
```

```python
vietFiles = glob.glob('./Cleaned/*Viet.txt')
```

```python
len(vietFiles)
```

```python
vietFiles[0]
```

```python
freq = {}
```

```python
def generateWordList(filePath):
    print("Working on " + filePath)
    with open(filePath, 'r') as infile:
        text = infile.read()
    words = nltk.word_tokenize(text)
    for word in words:
        if (word not in string.punctuation): #not a punctuation
            if (word.lower() not in freq): #not in the dict yet
                freq[word.lower()] = 0
            else:
                freq[word.lower()] = freq[word.lower()] + 1            
```

```python
for file in vietFiles:
    generateWordList(file)
```

```python
freq
```

```python
newFreq = {}
for (key, value) in freq.items():
    if (value >= 2): #word appears less than 2 times
        newFreq[key] = value
    else:
        print("Deleting " + key) 
```

```python
with open('freqlist.json', 'w') as outfile:
    json.dump(newFreq, outfile)
```
